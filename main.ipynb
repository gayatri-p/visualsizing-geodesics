```python
from solver import *
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'notebook', 'no-latex'])
plt.rcParams.update({"font.size":12}) 
```

```python
E=[0.8, 1, 1.1, 1.2, 1.5][0]
Lz=5
M=1
epsilon=0
schwarz = Schwarzschild(M = M)

# Initial Conditions
t0 = 0
r0 = 10
theta0 = np.pi/2
phi0 = 0

# p0 = 0 # time
# p1 = -1 # r
p2 = 0 # theta
# p3 = 0 # phi

n_rays = 1 # number of light rays
N = 500 # number of simulation points

lines = []
for i in range(1, n_rays+1):
    p0, p1, p3 = schwarz.compute_4momentum(r0, E, Lz, epsilon)
    y0 = [t0, r0, theta0, phi0, p0, p1, p2, p3]

    # Solve the equations of motion for a specified number of steps with a specified initial step size
    tau, sol = schwarz.solve(y0, N, 1e-5)

    t, r, theta, phi = sol[:, 0], sol[:, 1], sol[:, 2], sol[:, 3]

    x = r*np.cos(phi)*np.sin(theta)
    y = r*np.sin(phi)*np.sin(theta)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))
ax.plot(x, y)
ax.set_xlim(-10, 10)
ax.set_ylim(-8, 8)
EH = plt.Circle((0, 0), schwarz.r_EH, color = "k", label='Black Hole $(R_\mathrm{EH} =$'+f'${schwarz.r_EH:.1f})$')
ax.add_patch(EH)
EH = plt.Circle((0, 0), 3*M, color = "red", linestyle='--', fill=False, label='Photosphere $(R_\mathrm{ph} =$'+f'${3*M:.1f})$')
ax.add_patch(EH)
ax.set_aspect("equal")
ax.text(0, -7, '(a) Massive particle')
```
